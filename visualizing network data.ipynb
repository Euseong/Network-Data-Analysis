## 3.3 Graph Layouts

```r
install.packages("sand")
install.packages("igraph")
```

```r
library(sand)
g.l <- graph.lattice(c(5, 5, 5))

options(warn=0)
data(aidsblog)
summary(aidsblog)
```

### adisblog graph가 구버전 igraph에서 만들어졌기 때문에 upgrade_grph() 함수를 사용하여 현재 버전으로 변환함

```r
aidsblog <- (upgrade_graph(aidsblog))
summary(aidsblog)
```

#### warning이 발생하지 않음

### circular layout

```r
igraph.options(vertex.size=3, vertex.label=NA, vertex.color="skyblue", edge.arrow.size=0.5)
par(mfrow=c(1, 2))
set.seed(100)
plot(g.l, layout=layout.circle)
title("5x5x5 Lattice")
plot(aidsblog, layout=layout.circle)
title("Blog Network")
par(mfrow=c(1, 1))
```

- igraph.options() 함수를 통해 network plot의 다양한 특성을 조정할 수 있음
- igraph.options()를 사용하지 않고 plot() 함수 내에서도 해당 옵션 부여 가능

```r
plot(g.l, layout=layout.circle, vertex.color="red")
```

## 3.4 Decorating Graph Layouts

- karate club network

```r
library(igraphdata)
data(karate)
set.seed(42)
l <- layout.kamada.kawai(karate)

igraph.options(vertex.size=10) # vertex size를 10으로 설정
par(mfrow=c(1,1))
plot(karate, layout=l, vertex.label=V(karate)) # karate 데이터의 vertex들을 kamada.kawai layout으로 plot
```

```r
V(karate)$name
```

```r
V(karate)$label <- sub("Actor ", "", V(karate)$name) # vertex 이름에서 "Actor "을 공백""으로 대체

V(karate)$shape <- "circle" # vertex 모양을 원형으로 변환
V(karate)[c("Mr Hi", "John A")]$shape <- "rectangle" # 이름이 "Mr Hi", "John A"인 vertex를 사각형 모양으로 변환
```

```r
V(karate)[Faction]
```

```r
# 각 vertex의 체육관 종류에 따라 다른 색 부여
V(karate)[Faction == 1]$color <- "red"
V(karate)[Faction == 2]$color <- "dodgerblue"
```

```r
V(karate)$size
```

```r
V(karate)$size2
```

```r
graph.strength(karate) # 각 vertex의 degree를 계산하는 함수
```

```r
# Vertex의 크기를 degree에 비례하게 변환
V(karate)$size <- 4*sqrt(graph.strength(karate))
V(karate)$size2 <- V(karate)$size * .5

# edge의 너비를 weight에 비례하게 변환
E(karate)$width <- E(karate)$weight
```

```r
# edge가 어떤 체육관 사이를 이어주는 지에 따라 다른 색 부여
F1 <- V(karate)[Faction==1]
F2 <- V(karate)[Faction==2]
E(karate)[ F1 %--% F1 ]$color <- "pink" # 체육관 1끼리 연결하는 edge는 pink
E(karate)[ F2 %--% F2 ]$color <- "lightblue" # 체육관 1과 2를 연결하는 edge는 lightblue
E(karate)[ F1 %--% F2 ]$color <- "yellow" # 체육관 2끼리 연결하는 edge는 yellow

# vertex 크기가 작으면 label을 vertex 안에 표기하기 어려우므로 label을 vertex에서 조금 떨어지게 위치
V(karate)$label.dist <- ifelse(V(karate)$size >= 10, 0, 0.75)

plot(karate, layout=l)
```

- Lazega lawyer network

```r
data(lazega)
lazega <- upgrade_graph(lazega)
plot(lazega, layout=layout.fruchterman.reingold)
```

```r
V(lazega)$Office
```

```r
# vertex의 Office 지역에 따라 다른 색 부여
colbar <- c("red", "dodgerblue", "goldenrod")
v.colors <- colbar[V(lazega)$Office]
```

```r
V(lazega)$Practice
```

```r
# vertex의 Practice에 따라 다른 모양 부여
v.shapes <- c("circle", "square")[V(lazega)$Practice]
```

```r
V(lazega)$Years
```

```r
# vertex의 Year에 따라 비례하는 크기 부여
v.size <- 3.5*sqrt(V(lazega)$Years)
```

```r
V(lazega)$Seniority
```

```r
# vertex의 Seniority를 label로
v.label <- V(lazega)$Seniority
# Reproducible layout.
set.seed(42)
l <- layout.fruchterman.reingold(lazega)
plot(lazega, layout=l, vertex.color=v.colors, vertex.shape=v.shapes, vertex.size=v.size, vertex.label=v.label)
```

## 3.5 Visualizing Large Networks

- French political blog network

```r
fblog <- upgrade_graph(fblog)
summary(fblog)
```

```r
party.names <- sort(unique(V(fblog)$PolParty))
party.names
```

```r
V(fblog)$PolParty
```

```r
set.seed(42)
l = layout.kamada.kawai(fblog) # kamada kawai layout
party.nums.f <- as.factor(V(fblog)$PolParty) # PolParty 특성을 factor로 변환
party.nums <- as.numeric(party.nums.f) # PolParty 값에 따라 다른 숫자 부여

# PolParty에 따라 다른 vertex 색 부여(vertex.color=party.nums)
plot(fblog, layout=l, vertex.label=NA, vertex.color=party.nums, vertex.size=3)
```

```r
set.seed(42)
l <- layout.drl(fblog) # drl layout으로 다시 plot
plot(fblog, layout=l, vertex.size=5, vertex.label=NA, vertex.color=party.nums)
```

```r
fblog.c <- contract.vertices(fblog, party.nums)
E(fblog.c)$weight <- 1
fblog.c <- simplify(fblog.c) # removes the loop and/or multiple edges from a graph
```

```r
party.size <- as.vector(table(V(fblog)$PolParty)) # 데이터에 같은 PolParty가 많을 수록 해당 PolParty의 크기가 크도록 설정

# edge의 너비가 weight에 비례하도록 설정하여 plot
plot(fblog.c, vertex.size=5*sqrt(party.size), vertex.label=party.names, vertex.color=V(fblog.c),
     edge.width=sqrt(E(fblog.c)$weight), vertex.label.dist=1.5, edge.arrow.size=0)
```

- karate network

```r
# find the vertices not farther than a given limit from another fixed vertex, i.e. neighbors
data(karate)
k.nbhds <- graph.neighborhood(karate, order=1)
```

```r
k.nbhds
```

```r
# number of neighbors for each vertex
sapply(k.nbhds, vcount)
```

```r
k.1 <- k.nbhds[[1]] # 1번째 vertex는 karate data의 1번째 vertex, 나머지는 그 vertex의 neighbor인 sub-graph
k.34 <- k.nbhds[[34]] # 34번째 vertex는 karate data의 34번째 vertex, 나머지는 그 vertex의 neighbor인 sub-graph

# neighbor를 구한 vertex의 색은 red, neighbor들은 lightblue로 하여 plot
par(mfrow=c(1,2))
plot(k.1, vertex.label=NA, vertex.color=c("red", rep("lightblue", 16)))
plot(k.34, vertex.label=NA, vertex.color=c(rep("lightblue", 17), "red"))
par(mfrow=c(1,1))
```
